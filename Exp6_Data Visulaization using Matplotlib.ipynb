```python
#importing the library
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=x*2
```

```python
y
```

# Line Chart

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

# Bar Chart

```python
plt.bar(x,y)
plt.bar(x,y, color="cyan")
plt.title("Bar Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

# Scatter Plot

```python
a=(1,4,6,3,8,2,9,5)
b=(10,33,42,23,17,15,21,64)
plt.scatter(a,b, color="cyan")
plt.scatter(a,b)
plt.title("scatter Plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
